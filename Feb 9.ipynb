```python
import pandas as pd
```

```python
df = pd.read_csv('Student-Performance-csv_T4qDx.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
#convert Extracurricular activities column
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()
le.fit_transform(df['Extracurricular Activities'])

df['Extracurricular Activities'] = le.fit_transform(df['Extracurricular Activities'])
df
```

```python
x = df[['Hours Studied',	'Previous Scores',	'Extracurricular Activities',	'Sleep Hours',	'Sample Question Papers Practiced']]
y = df['Performance Index']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
#transforming the distribution of dataset - standarize the dataset

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
import pickle 
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
lr = LinearRegression()

lr.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test_scaled)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
#save the model
with open('student_lr_final_model.pkl', 'wb') as file:
    pickle.dump((lr,scaler,le),file)
    #here the pickle is storing the fitted model - lr, but also we are storing other variables that we used to do preprocessing
    #so that when we run the model again, it will know what to do in terms of preprocessing by using the variable scaler and le
```
